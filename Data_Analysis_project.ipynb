```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_url = 'https://raw.githubusercontent.com/datastray/python_project/main/sales_data.csv'
df = pd.read_csv(data_url)

# View first 5 rows of the dataset
df.head()
```

```python
df.info()
```

```python
product_category_revenue = df.groupby('product_category')['total_revenue'].sum()
product_category_revenue
total_category_revenue = df['total_revenue'].sum()
product_category_percentage = (product_category_revenue / total_category_revenue * 100)
product_category_percentage
```

```python
product_category_percentage.plot(kind='bar')
plt.title('Percentage of Total Revenue by Product Category')
plt.ylabel('Percentage (%)')
plt.show()
```

```python
product_location_revenue = df.groupby(['product_category', 'location'])['total_revenue'].sum()
product_location_revenue
```
